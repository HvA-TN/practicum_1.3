```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1.33 # water (assume n1=1, n=n2 of water)
L = 5e-2 # meter

thetai = np.linspace(0,90,100)
thetai = np.deg2rad(thetai)
d = L*np.tan(np.arcsin((1/n)*np.sin(thetai)))
d[d>L] = np.nan
dy = d*np.sin(thetai*(1-1/n))

plt.figure(dpi=300)
plt.plot(np.rad2deg(thetai), dy*1000, 'r.-')
plt.grid()
plt.xlabel(r"$\theta_i$ $[^\circ]$")
plt.ylabel(r"$\Delta y$ [mm]")
```

```python
n = np.linspace(1.3, 1.6,100) # water (assume n1=1, n=n2 of water)
thetai = 70 # degree
L = 5e-2 # meter

thetai = np.deg2rad(thetai)
d = L*np.tan(np.arcsin((1/n)*np.sin(thetai)))
#d[d>L] = np.nan
dy = d*np.sin(thetai*(1-1/n))

plt.figure(dpi=300)
plt.plot(n, dy*1000, 'r.-')
plt.grid()
plt.xlabel(r"$n$ $[-]$")
plt.ylabel(r"$\Delta y$ [mm]")
```
